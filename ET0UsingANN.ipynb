```python
from keras.callbacks import ModelCheckpoint
from keras.models import Sequential
from keras.layers import Dense, Activation, Flatten
from sklearn.model_selection import train_test_split
#from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error 
from matplotlib import pyplot as plt
import seaborn as sb
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import datetime as dt
import os
#import warnings 
#warnings.filterwarnings('ignore')
#warnings.filterwarnings('ignore', category=DeprecationWarning)
#from xgboost import XGBRegressor
```

```python
dataDir = 'C:/Users/ADMIN/Desktop/Notebooks/Work/DataSet/DataForET0'
trainX = pd.read_csv(os.path.join(dataDir,'TrainValidData.csv')).values[:,:6]
trainY = pd.read_csv(os.path.join(dataDir,'TrainValidData.csv')).values[:,6]
testX = pd.read_csv(os.path.join(dataDir,'TestData.csv')).values[:,:6]
testY = pd.read_csv(os.path.join(dataDir,'TestData.csv')).values[:,6]
t=dt.date.today()
print(t)
```

```python
NN_model = Sequential()
NN_model.add(Dense(20, kernel_initializer='normal',input_dim = trainX.shape[1], activation='relu'))
#NN_model.add(Dense(256, kernel_initializer='normal',activation='relu'))
#NN_model.add(Dense(18, kernel_initializer='normal',activation='relu'))
NN_model.add(Dense(256, kernel_initializer='normal',activation='relu'))
NN_model.add(Dense(1, kernel_initializer='normal',activation='linear'))
NN_model.compile(loss='mean_absolute_error', optimizer='adam', metrics=['mean_absolute_error'])
#NN_model.summary()

checkpoint_name = 'ModelsForDifferentWeights/Weights-{epoch:03d}--{val_loss:.5f}.hdf5' 
checkpoint = ModelCheckpoint(checkpoint_name, monitor='val_loss', verbose = 1, save_best_only = True, mode ='auto')
callbacks_list = [checkpoint]
```

```python
NN_model.fit(trainX, trainY, epochs=200, batch_size=32, validation_split = 0.2, callbacks=callbacks_list)
```

```python
# Load wights file of the best model :
weights_file = 'ModelsForDifferentWeights/Weights-466--0.09921.hdf5' # choose the best checkpoint 
#weights_file = 'Backup/Models/Weights-496--0.09607.hdf5' # choose the best checkpoint 
NN_model.load_weights(weights_file) # load it
```

```python
NN_model.compile(loss='mean_absolute_error', optimizer='adam', metrics=['mean_absolute_error'])
```

```python
#using on the test data set
predictions = NN_model.predict(testX)
np.savetxt('PredictedValues.csv', predictions, header='Predicted values', delimiter=',')
testY.resize(len(testY),1)
pred=pd.read_csv('PredictedValues.csv').values
error=predictions-testY
error_mse=(((testY-predictions)/testY)**2)*100
#print(abs(error))
print(max(error))
#max(error_mse)
np.savetxt('Error.csv', error, header='Error', delimiter=',')
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
fig = plt.figure()
ax = plt.axes()
#number of samples
num=100
x = np.linspace(0, num, num)
plt.plot(x, pred[:num])
plt.plot(x, testY[:num])
plt.legend(('ANN model estimate', 'PM'))
plt.xlabel('Number of sample outputs')
plt.ylabel('Reference evapotranspiration (mm/day)')
plt.title('Evapotranspiration comparison: PM vs ANN estimation')
plt.savefig('et0Comparison.jpg')
```

```python
num=100
x = np.linspace(0, num, num)
plt.errorbar(x, error[:num], yerr=0.05, fmt='.k', 
             color='black', ecolor='lightgray', elinewidth=2, 
             capsize=0, barsabove='true');
plt.xlabel('Number of sample outputs')
plt.ylabel('Error in estimated output (mm/day)')
plt.savefig('errorInEt0.jpg')
```
